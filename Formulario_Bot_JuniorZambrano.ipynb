```python
#librerias necesarias
import pandas as pd
import warnings
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
```

```python
#lineas para ignnorar advertencias 
warnings.filterwarnings('ignore')
warnings.warn('DelftStack')
warnings.warn('Do not show this message')
```

```python
#Lectura de dataframe
df = pd.read_csv('data.csv')
df
```

```python
#Diccionario 

S_dict = {'Si':'//*[@id="i22"]/div[3]/div',
          'No':'//*[@id="i25"]/div[3]/div'}
#Diccionario 

SP_dict = {'Primero':'//*[@id="i32"]/div[3]/div',
           'Segundo':'//*[@id="i35"]/div[3]/div',
           'Tercero':'//*[@id="i38"]/div[3]/div',
           'Cuarto':'//*[@id="i41"]/div[3]/div',
           'Quinto':'//*[@id="i44"]/div[3]/div',
           'Sexto':'//*[@id="i47"]/div[3]/div',
           'Septimo':'//*[@id="i50"]/div[3]/div',
           'Octavo':'//*[@id="i53"]/div[3]/div'}

for row, datos in df.iterrows():
    Nombres = datos['Nombres']
    Apellidos = datos['Apellidos']
    CI = datos['Cédula de Identidad']
    FN = datos['Fecha de Nacimiento']
    CS = datos['¿Posee conocimiento de Seguridad?']
    SP = datos['Semestre al que pertenece']
    
    #Driver y dirección
    options = Options()
    w = webdriver.Chrome(executable_path='chromedriver.exe', chrome_options=options)
    w.get("https://forms.gle/1vPpQGqL4Qm1AwdT7")
    
    #nombre
    WebDriverWait(w, 20).until(EC.element_to_be_clickable((By.XPATH, f'//*[@id="mG61Hd"]/div[2]/div/div[2]/div[1]/div/div/div[2]/div/div[1]/div/div[1]/input'))).send_keys(Nombres)
    
    #apellido
    WebDriverWait(w, 20).until(EC.element_to_be_clickable((By.XPATH, f'//*[@id="mG61Hd"]/div[2]/div/div[2]/div[2]/div/div/div[2]/div/div[1]/div/div[1]/input'))).send_keys(Apellidos)
    
    #cedula
    WebDriverWait(w, 20).until(EC.element_to_be_clickable((By.XPATH, f'//*[@id="mG61Hd"]/div[2]/div/div[2]/div[3]/div/div/div[2]/div/div[1]/div/div[1]/input'))).send_keys(CI)
   
    #Fecha de Nacimiento
    WebDriverWait(w, 20).until(EC.element_to_be_clickable((By.XPATH, f'//*[@id="mG61Hd"]/div[2]/div/div[2]/div[4]/div/div/div[2]/div/div/div[2]/div[1]/div/div[1]/input'))).send_keys(FN)

    #¿Posee conocimiento de Seguridad?
    S = CS
    WebDriverWait(w, 20).until(EC.element_to_be_clickable((By.XPATH, S_dict[S]))).click()

    #Semestre al que pertenece
    SP = SP
    WebDriverWait(w, 20).until(EC.element_to_be_clickable((By.XPATH, SP_dict[SP]))).click()
    
    #Boton
    WebDriverWait(w, 20).until(EC.element_to_be_clickable((By.XPATH, f'//*[@id="mG61Hd"]/div[2]/div/div[3]/div[1]/div[1]/div/span/span'))).click()
```
